You should end up with an empty project window like that shown in figure 4. To the right you'll see a catalog of "parts" that you can drag onto the schematic. Under the Analog -> DAC folder grab a "Waveform DAC (8-bit)" and drag it into the schematic. The term "DAC" stands for "Digital to Analog Converter". This particular tool takes a stored sequence of 8-bit digital values (0-255) and converts them to analog voltages that are periodically applied to the output to produce a pre-defined waveform. In other words, it's like a crude version of the function generator on your bench built in to the PSoC for your use. After you drag the WaveDAC onto the schematic you'll see that you immediately get two "Errors" and one "Note". These are telling you that the project as it is will not build. This is because (1) a required digital input (ws) is not present and (2) the analog output of the DAC doesn't go anywhere.

Let's fix those:

1. From the Digital->Logic folder grab a "Logic Low '0'" and attach it to the 'ws' input
2. From the Ports and Pins folder get an "Analog Pin" and connect the output to that.

You'll need to "right-click -> Shape -> Flip Horizontal" to get the Pin oriented correctly. Use the "wiring tool" to connect the logic low to the 'ws' input and the analog pin to the output of the WaveDAC. When you've got things wired correclty it should look like Fig. 7:

After the build is complete you can program the device (see arrow in Fig. 8). (You may need to update the firmware on the device depending on whether it's been updated recently). You can use the "stand alone programmer" application to do this.

After the device is programmed you should see a screen like Fig. 9. Click on the "Pins" configuration (see the arrow indicating where to double-click). You'll see where the router assigned the output pin (Pin 1) on your device. Mine ended up on Pin 15.1.

Now, use an oscilloscope to measure the voltage on the output pin and confirm that the device was programmed correclty.

Now build the rest of the circuit! Your instructor will give you guidance and motivation. Basically the PGA is a built in amplifier that we'll be using to boost the output voltage from the WaveDAC. The comparator will turn this into a digital waveform, and the flip flops will act as a "divide by 8" counter. In this particular example we have no real need for these functions, this time, but we need to understand the building blocks before we can actually build something!
